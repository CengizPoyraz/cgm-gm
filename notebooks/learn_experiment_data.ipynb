```python
from obspy import read
import matplotlib.pyplot as plt
import numpy as np
# File path provided by the user
file_path = "/home/developer/Desktop/NOTEBOOK/GMS/cgm-gm/PhaseNet/06/07/ADVT_20230206_0700_100_B_KO_E.GCF"

# Read the file to extract trace data
stream = read(file_path)
trace = stream[0]  # Assuming a single trace in the stream

# Extract trace data and create time array
trace_data = trace.data  # Amplitude data as a NumPy array
sampling_rate = trace.stats.sampling_rate
npts = trace.stats.npts
duration = npts / sampling_rate
time_array = np.linspace(0, duration, npts)

# Plot the seismic data
plt.figure(figsize=(12, 6))
plt.plot(time_array, trace_data, linewidth=0.5)
plt.title("Seismic Trace Data")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
